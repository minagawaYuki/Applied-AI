```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('Train Dataset.csv')
df_cleaned = df[~df['Gender'].str.contains(r'\d', na=False)]
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
sct1 = pd.DataFrame(data = df)
sct1.plot.scatter(x='Daily_Usage_Time (minutes)', y='Posts_Per_Day', c='blue', ax=ax)

x = df['Daily_Usage_Time (minutes)']
y = df['Posts_Per_Day']

slope, intercept = np.polyfit(x, y, 1)

regression_line = slope * x + intercept

ax.plot(x, regression_line, color='red', label='Regression Line')

ax.set_title('Daily Usage Time vs Posts Per Day')
ax.set_xlabel('Daily Usage Time (minutes)')
ax.set_ylabel('Posts Per Day')

ax.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Check if the DataFrame is populated and valid
if not df.empty:
    df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
    df['Daily_Usage_Time (minutes)'] = pd.to_numeric(df['Daily_Usage_Time (minutes)'], errors='coerce')

    df = df.dropna(subset=['Age', 'Daily_Usage_Time (minutes)'])

    fig, ax = plt.subplots()

    df.plot.scatter(x='Age', y='Daily_Usage_Time (minutes)', color='blue', ax=ax)

    x = df['Age']
    y = df['Daily_Usage_Time (minutes)']

    slope, intercept = np.polyfit(x, y, 1)

    regression_line = slope * x + intercept

    ax.plot(x, regression_line, color='red', label=f'Regression Line (slope={slope:.2f})')

    ax.set_title('Age vs Daily Usage Time')
    ax.set_xlabel('Age')
    ax.set_ylabel('Daily Usage Time (minutes)')

    ax.grid(True)

    ax.legend()

    plt.show()
else:
    print("DataFrame is empty.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Train Dataset.csv')

grouped_emotion_usage = df.groupby('Dominant_Emotion')['Daily_Usage_Time (minutes)'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(grouped_emotion_usage['Dominant_Emotion'], grouped_emotion_usage['Daily_Usage_Time (minutes)'], color='coral')

plt.title('Average Daily Usage Time by Dominant Emotion')
plt.xlabel('Dominant Emotion')
plt.ylabel('Average Daily Usage Time (minutes)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_cleaned = df[~df['Gender'].str.contains(r'\d', na=False)]

grouped_emotion_gender_usage = df_cleaned.groupby(['Gender', 'Dominant_Emotion'])['Daily_Usage_Time (minutes)'].mean().unstack()

grouped_emotion_gender_usage.plot(kind='bar', figsize=(10, 6))

plt.title('Average Daily Usage Time by Dominant Emotion and Gender')
plt.xlabel('Gender')
plt.ylabel('Average Daily Usage Time (minutes)')
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```
