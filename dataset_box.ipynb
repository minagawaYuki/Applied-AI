```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

print("Libraries installed successfully")
```

```python
path = "C:/Users/Charuzu/OneDrive/Documents/3rd Year/Applied AI/Midterm/train.csv"
df = pd.read_csv(path)

# Drop rows with missing values 
df_cleaned = df.dropna()

# Remove any leading/trailing spaces
df['Gender'] = df['Gender'].str.strip()

# Remove duplicates 
df_cleaned = df.drop_duplicates()

# Example: Removing rows with invalid Gender
df = df[df['Gender'].isin(['Male', 'Female', 'Non-binary', 'Other'])]
```
