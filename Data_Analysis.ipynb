```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = "C:/Users/pcx/Documents/SCHOOL 3RD YEAR/CSIT349/Midterm Project/Train Dataset.csv"
df = pd.read_csv(path)
```

```python
print(df.info())
print(df.describe())
```

```python
#For the Mean and Median
# DUT = Daily_Usage_Time (minutes)
# PPD = Posts_Per_Day
# LRPD = Likes_Received_Per_Day
# CRPD = Comments_Received_Per_Day
# MSPD = Messages_Sent_Per_Day

# Mean
Mean_Age = df['Age'].mean()
Mean_DUT = df['Daily_Usage_Time (minutes)'].mean()
Mean_PPD = df['Posts_Per_Day'].mean()
Mean_LRPD = df['Likes_Received_Per_Day'].mean()
Mean_CRPD = df['Comments_Received_Per_Day'].mean()
Mean_MSPD = df['Messages_Sent_Per_Day'].mean()

# Median
Median_Age = df['Age'].median()
Median_DUT = df['Daily_Usage_Time (minutes)'].median()
Median_PPD = df['Posts_Per_Day'].median()
Median_LRPD = df['Likes_Received_Per_Day'].median()
Median_CRPD = df['Comments_Received_Per_Day'].median()
Median_MSPD = df['Messages_Sent_Per_Day'].median()

# Mean
print("Mean for Age: " + str(Mean_Age))
print("Mean for Daily_Usage_Time (minutes): " + str(Mean_DUT))
print("Mean for Posts_Per_Day: " + str(Mean_PPD))
print("Mean for Likes_Received_Per_Day: " + str(Mean_LRPD))
print("Mean for Comments_Received_Per_Day: " + str(Mean_CRPD))
print("Mean for Messages_Sent_Per_Day: " + str(Mean_MSPD))
print("\n")
# Median
print("Median for Age: " + str(Median_Age))
print("Median for Daily_Usage_Time: " + str(Median_DUT))
print("Median for Posts_Per_Day: " + str(Median_PPD))
print("Median for Likes_Received_Per_Day: " + str(Median_LRPD))
print("Median for Comments_Received_Per_Day: " + str(Median_CRPD))
print("Median for Messages_Sent_Per_Day: " + str(Median_MSPD))



```

```python
# Mode
Mode_Age = df['Age'].mode().iloc[0]
Mode_DUT = df['Daily_Usage_Time (minutes)'].mode().iloc[0]
Mode_PPD = df['Posts_Per_Day'].mode().iloc[0]
Mode_LRPD = df['Likes_Received_Per_Day'].mode().iloc[0]
Mode_CRPD = df['Comments_Received_Per_Day'].mode().iloc[0]
Mode_MSPD = df['Messages_Sent_Per_Day'].mode().iloc[0]

print("Mode for Age: " + str(Mode_Age))
print("Mode for Daily_Usage_Time: " + str(Mode_DUT))
print("Mode for Posts_Per_Day: " + str(Mode_PPD))
print("Mode for Likes_Received_Per_Day: " + str(Mode_LRPD))
print("Mode for Comments_Received_Per_Day: " + str(Mode_CRPD))
print("Mode for Messages_Sent_Per_Day: " + str(Mode_MSPD))
```

```python
# Standard Deviation and Variance
Std_Age = df['Age'].std()
Std_DUT = df['Daily_Usage_Time (minutes)'].std()
Std_PPD = df['Posts_Per_Day'].std()
Std_LRPD = df['Likes_Received_Per_Day'].std()
Std_CRPD = df['Comments_Received_Per_Day'].std()
Std_MSPD = df['Messages_Sent_Per_Day'].std()

Var_Age = df['Age'].var()
Var_DUT = df['Daily_Usage_Time (minutes)'].var()
Var_PPD = df['Posts_Per_Day'].var()
Var_LRPD = df['Likes_Received_Per_Day'].var()
Var_CRPD = df['Comments_Received_Per_Day'].var()
Var_MSPD = df['Messages_Sent_Per_Day'].var()

print("Standard Deviation for Age: " + str(Std_Age))
print("Standard Deviation for Daily_Usage_Time: " + str(Std_DUT))
print("Standard Deviation for Posts_Per_Day: " + str(Std_PPD))
print("Standard Deviation for Likes_Received_Per_Day: " + str(Std_LRPD))
print("Standard Deviation for Comments_Received_Per_Day: " + str(Std_CRPD))
print("Standard Deviation for Messages_Sent_Per_Day: " + str(Std_MSPD))
print("\n")
print("Variance Value for Age: " + str(Var_Age))
print("Variance Value for Daily_Usage_Time: " + str(Var_DUT))
print("Variance Value for Posts_Per_Day: " + str(Var_PPD))
print("Variance Value for Likes_Received_Per_Day: " + str(Var_LRPD))
print("Variance Value for Comments_Received_Per_Day: " + str(Var_CRPD))
print("Variance Value for Messages_Sent_Per_Day: " + str(Var_MSPD))
```

```python
# For Min, Max, and Range

# MIN
Min_Age = df['Age'].min()
Min_DUT = df['Daily_Usage_Time (minutes)'].min()
Min_PPD = df['Posts_Per_Day'].min()
Min_LRPD = df['Likes_Received_Per_Day'].min()
Min_CRPD = df['Comments_Received_Per_Day'].min()
Min_MSPD = df['Messages_Sent_Per_Day'].min()

# MAX
Max_Age = df['Age'].max()
Max_DUT = df['Daily_Usage_Time (minutes)'].max()
Max_PPD = df['Posts_Per_Day'].max()
Max_LRPD = df['Likes_Received_Per_Day'].max()
Max_CRPD = df['Comments_Received_Per_Day'].max()
Max_MSPD = df['Messages_Sent_Per_Day'].max()

# RANGE
Range_Age = Max_Age - Min_Age
Range_DUT = Max_DUT - Min_DUT
Range_PPD = Max_PPD - Min_PPD
Range_LRPD = Max_LRPD - Min_LRPD
Range_CRPD = Max_CRPD - Min_CRPD
Range_MSPD = Max_MSPD - Min_MSPD


print("Minimum for Age: " + str(Min_Age))
print("Minimum for Daily_Usage_Time: " + str(Min_DUT))
print("Minimum for Posts_Per_Day: " + str(Min_PPD))
print("Minimum for Likes_Received_Per_Day: " + str(Min_LRPD))
print("Minimum for Comments_Received_Per_Day: " + str(Min_CRPD))
print("Minimum for Messages_Sent_Per_Day: " + str(Min_MSPD))
print("\n")
print("Maximum for Age: " + str(Max_Age))
print("Maximum for Daily_Usage_Time: " + str(Max_DUT))
print("Maximum for Posts_Per_Day: " + str(Max_PPD))
print("Maximum for Likes_Received_Per_Day: " + str(Max_LRPD))
print("Maximum for Comments_Received_Per_Day: " + str(Max_CRPD))
print("Maximum for Messages_Sent_Per_Day: " + str(Max_MSPD))
print("\n")
print("Range for Age: " + str(Range_Age))
print("Range for Daily_Usage_Time: " + str(Range_DUT))
print("Range for Posts_Per_Day: " + str(Range_PPD))
print("Range for Likes_Received_Per_Day: " + str(Range_LRPD))
print("Range for Comments_Received_Per_Day: " + str(Range_CRPD))
print("Range for Messages_Sent_Per_Day: " + str(Range_MSPD))

```
